```julia
using JLD2
using Plots
# plotlyjs()
```

```julia
dSmooth = load("$(@__DIR__)/data/efficiencyResults1.jld2");
timingsSmooth = dSmooth["timings"];
errorsSmooth = dSmooth["errors"];
AlgNamesSmooth = dSmooth["AlgNames"];
```

```julia
for x in AlgNamesSmooth
    println(x)
end
```

```julia
# Plot efficiency of algorithms
p1 = plot(timingsSmooth[:, 1], errorsSmooth[:, 1], label="Upwind: order 1", xlabel="Time", ylabel="Relative error", yscale=:log10, xscale=:log10, ls=:dashdot, markershape=:circle, markersize=3, legend=:bottomleft, c=:blue)
plot!(p1, timingsSmooth[:, 2], errorsSmooth[:, 2], label="WENO: order 2", ls=:dashdotdot, markershape=:circle, markersize=3, c=:green)
plot!(p1, timingsSmooth[:, 3], errorsSmooth[:, 3], label="Upwind + MOOD: order 2", ls=:dashdot, markershape=:star, markersize=3, c=:purple)
plot!(p1, timingsSmooth[:, 6], errorsSmooth[:, 6], label="MUSCL: order 2", ls=:solid, markershape=:circle, markersize=3, c=:gold3)
plot!(p1, timingsSmooth[:, 4], errorsSmooth[:, 4], label="MUSCL + MOOD: order 2", ls=:solid, markershape=:star, markersize=3, c=:gold3)
plot!(p1, timingsSmooth[:, 7], errorsSmooth[:, 7], label="MUSCL: order 4", ls=:solid, markershape=:circle, markersize=3, c=:maroon1)
plot!(p1, timingsSmooth[:, 5], errorsSmooth[:, 5], label="MUSCL + MOOD: order 4", ls=:solid, markershape=:star, markersize=3, c=:maroon1)
display(p1)
savefig(p1, "$(@__DIR__)/efficiency_smoothInit.pdf")
```
